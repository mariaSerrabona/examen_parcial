Unir los datos de navegación ya tratados con los datos de conversiones,creando una columna de 0 y 1 indicando si el usuario no ha convertido o si ha convertido.
```python
#creamos una nueva columna en conv 
def columna_conversion(df):
    for i in df:
        conversion = '1'
    df['conversion'] = conversion
```

```python
def columna_datos_dummies(df):
    df.loc[df['conversion'].isna(), 'conversion']=0
```

```python
def crear_df_total(df1,df2):
    columna_conversion(df1)
    df_total=pd.concat([df2 , df1])
    columna_datos_dummies(df_total)
    return df_total
```
